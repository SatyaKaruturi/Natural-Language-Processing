```python
from newspaper import Article
url = "https://www.latimes.com/world-nation/story/2019-12-12/les-miserables-why-are-the-french-who-seem-to-have-much-so-quick-to-protest"

Text_news = Article(url)
```

### Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize

tokenized_text = sent_tokenize(Text_news.text)
```

```python
print(tokenized_text)
```

### Word Tokenization

```python
from nltk.tokenize import word_tokenize

tokenized_word = word_tokenize(Text_news.text)
```

```python
print(tokenized_word)
```

### Frequency Distribution

```python
from nltk.probability import FreqDist

fdist_new = FreqDist(tokenized_word)

print(fdist_new)
```

```python
fdist_new.most_common(20)
```

```python
# Frequency Distribution Plot

import matplotlib.pyplot as plt

fdist_new.plot(30,cumulative=False)
plt.show()
```

### Stopwords

```python
from nltk.corpus import stopwords

stop_words=set(stopwords.words("english"))

print(stop_words)
```

```python
filtered_words=[]
```

```python
for w in tokenized_word:
    if w not in stop_words:
        filtered_words.append(w)
        
print("Tokenized words:",tokenized_word)
print("Filterd words:",filtered_words)
```

```python
fdist_new1 = FreqDist(filtered_words)
```

```python
fdist_new1.most_common(20)
```

```python
# Frequency Distribution Plot

import matplotlib.pyplot as plt

fdist_new1.plot(30,cumulative=False)
plt.show()
```

### Stemming

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()

stemmed_words=[]

for w in filtered_words:
    stemmed_words.append(ps.stem(w))

print("Filtered words:",filtered_words)
print("Stemmed words:",stemmed_words)
```

```python
fdist_new_stemmed = FreqDist(stemmed_words)
```

```python
fdist_new_stemmed.most_common(20)
```

```python
# Frequency Distribution Plot

import matplotlib.pyplot as plt

fdist_new_stemmed.plot(30,cumulative=False)
plt.show()
```
